Only for Maharashtra

```python
from matplotlib import pyplot as plt
```

```python
mh_df = df[df['STATES'] == 'Maharashtra'][['PRODUCT_NAME','PRODUCT_CATEGORY','PRICE', 'STATES']]
```

```python
mh_df
```

```python
mh_df['PRICE'] = mh_df['PRICE'].str.replace(",","").astype('float').astype('int')
```

```python
plt.hist(mh_df['PRICE'], edgecolor='black', color='blue', bins=50)
plt.title("Price count for Maharashtra State")
plt.xlabel('PRICES')
plt.ylabel("Count")
```

```python
mh_df['PRODUCT_CATEGORY'] = mh_df['PRODUCT_CATEGORY'].apply(lambda x : x.strip())
```

```python
mh_df['PRODUCT_CATEGORY'].value_counts()
```

```python
plt.hist(mh_df['PRODUCT_CATEGORY'], edgecolor='black', color='blue')
```
